```python
import pygame, sys
from pygame.locals import *
import random, time

# Initializing
pygame.init()

# Setting up FPS
FPS = 60
FramePerSec = pygame.time.Clock()

# Colors
BLUE  = (0, 0, 255)
RED   = (255, 0, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Game variables
SCREEN_WIDTH = 400
SCREEN_HEIGHT = 600
SPEED = 5
SCORE = 0

# Setting up fonts
font = pygame.font.SysFont("Verdana", 60)
font_small = pygame.font.SysFont("Verdana", 20)
game_over = font.render("Game Over", True, BLACK)

background = pygame.image.load("AnimatedStreet.png")

# Create game window
DISPLAYSURF = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
DISPLAYSURF.fill(WHITE)
pygame.display.set_caption("Game")

class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        self.image = pygame.image.load("Enemy.png")
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH-40), 0)  

    def move(self):
        global SPEED
        self.rect.move_ip(0, SPEED)
        if self.rect.top > SCREEN_HEIGHT:
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        self.image = pygame.image.load("Player.png")
        self.rect = self.image.get_rect()
        self.rect.center = (160, 520)

    def move(self):
        pressed_keys = pygame.key.get_pressed()
        if pressed_keys[K_LEFT] and self.rect.left > 0:
            self.rect.move_ip(-5, 0)
        if pressed_keys[K_RIGHT] and self.rect.right < SCREEN_WIDTH:
            self.rect.move_ip(5, 0)

class Coin(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.value = random.choice([1, 2, 3])  # Монеты могут стоить 1, 2 или 3 очка
        self.image = pygame.image.load("coin.png")
        size = 20 + (self.value * 5)  # Увеличиваем размер в зависимости от ценности
        self.image = pygame.transform.scale(self.image, (size, size))
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

    def move(self):
        self.rect.move_ip(0, 2)
        if self.rect.top > SCREEN_HEIGHT:
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

# Setting up sprites        
P1 = Player()
E1 = Enemy()

# Creating sprite groups
enemies = pygame.sprite.Group()
enemies.add(E1)
all_sprites = pygame.sprite.Group()
all_sprites.add(P1, E1)

coins = pygame.sprite.Group()
for _ in range(3):
    coins.add(Coin())

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == QUIT:
            running = False

    DISPLAYSURF.blit(background, (0, 0))
    scores = font_small.render(f"Coins: {SCORE}", True, BLACK)
    DISPLAYSURF.blit(scores, (10, 10))

    # Move and redraw all sprites
    for entity in all_sprites:
        DISPLAYSURF.blit(entity.image, entity.rect)
        entity.move()

    for coin in coins:
        DISPLAYSURF.blit(coin.image, coin.rect)
        coin.move()

    # Check if player collects a coin
    collected_coins = pygame.sprite.spritecollide(P1, coins, True)
    for coin in collected_coins:
        SCORE += coin.value  # Добавляем ценность монеты к общему счёту
        SPEED += 0.2  # Увеличиваем скорость врага за каждую собранную монету

    # Добавляем новые монеты, если их меньше 3
    while len(coins) < 3:
        coins.add(Coin())

    # Check collision with enemy
    if pygame.sprite.spritecollideany(P1, enemies):
        pygame.mixer.Sound('crash.wav').play()
        time.sleep(0.5)

        DISPLAYSURF.fill(RED)
        DISPLAYSURF.blit(game_over, (30, 250))
        pygame.display.update()

        time.sleep(2)
        running = False  # Завершаем игровой цикл

    pygame.display.update()
    FramePerSec.tick(FPS)

pygame.quit()
print("Game exited successfully.")
```

```python
import pygame, sys, time, random

# Difficulty
difficulty = 10

level = 1  # Starting level
score_to_next_level = 4  # Points needed to level up

# Window size
frame_size_x = 720
frame_size_y = 480

# Initialize pygame
pygame.init()
pygame.display.set_caption('Snake')
game_window = pygame.display.set_mode((frame_size_x, frame_size_y))

# Define colors
black = pygame.Color(0, 0, 0)
white = pygame.Color(255, 255, 255)
red = pygame.Color(255, 0, 0)
green = pygame.Color(0, 255, 0)

# FPS controller
fps_controller = pygame.time.Clock()

# Snake initial position and body
snake_pos = [100, 50]
snake_body = [[100, 50], [90, 50], [80, 50]]

# Function to generate food position and weight
def spawn_food():
    while True:
        food_x = random.randint(0, (frame_size_x // 10) - 1) * 10
        food_y = random.randint(0, (frame_size_y // 10) - 1) * 10
        if [food_x, food_y] not in snake_body:
            food_weight = random.choice([1, 2, 3])  # Food weight (1, 2, or 3 points)
            return [food_x, food_y], food_weight

food_pos, food_weight = spawn_food()
food_timer = 300  # Time before food disappears (in frames)

direction = 'RIGHT'
change_to = direction
score = 0
running = True  # Флаг для работы игры

# Function to handle game over
def game_over():
    global running
    my_font = pygame.font.SysFont('times new roman', 90)
    game_over_surface = my_font.render('Game Over', True, red)
    game_over_rect = game_over_surface.get_rect(midtop=(frame_size_x/2, frame_size_y/4))
    game_window.fill(black)
    game_window.blit(game_over_surface, game_over_rect)
    show_score_level()
    pygame.display.flip()
    time.sleep(3)
    running = False  # Завершаем игровой цикл

# Display score and level
def show_score_level():
    score_font = pygame.font.SysFont('consolas', 20)
    score_surface = score_font.render(f'Score: {score}  Level: {level}', True, white)
    score_rect = score_surface.get_rect(midtop=(frame_size_x/2, 15))
    game_window.blit(score_surface, score_rect)

# Main game loop
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and direction != 'DOWN':
                change_to = 'UP'
            if event.key == pygame.K_DOWN and direction != 'UP':
                change_to = 'DOWN'
            if event.key == pygame.K_LEFT and direction != 'RIGHT':
                change_to = 'LEFT'
            if event.key == pygame.K_RIGHT and direction != 'LEFT':
                change_to = 'RIGHT'
            if event.key == pygame.K_ESCAPE:
                running = False

    # Update direction
    direction = change_to

    # Move the snake
    if direction == 'UP':
        snake_pos[1] -= 10
    elif direction == 'DOWN':
        snake_pos[1] += 10
    elif direction == 'LEFT':
        snake_pos[0] -= 10
    elif direction == 'RIGHT':
        snake_pos[0] += 10

    # Snake growth logic
    snake_body.insert(0, list(snake_pos))
    if snake_pos == food_pos:
        score += food_weight
        food_pos, food_weight = spawn_food()
        food_timer = 300  # Reset food timer
        if score % score_to_next_level == 0:
            level += 1
            difficulty += 5
    else:
        snake_body.pop()

    # Decrease food timer and respawn food if timer expires
    food_timer -= 1
    if food_timer <= 0:
        food_pos, food_weight = spawn_food()
        food_timer = 300  # Reset timer for new food

    # Check for collisions with walls
    if snake_pos[0] < 0 or snake_pos[0] >= frame_size_x or snake_pos[1] < 0 or snake_pos[1] >= frame_size_y:
        game_over()

    # Check for collisions with itself
    if snake_pos in snake_body[1:]:
        game_over()

    # Draw game elements
    game_window.fill(black)
    for pos in snake_body:
        pygame.draw.rect(game_window, green, pygame.Rect(pos[0], pos[1], 10, 10))
    
    # Draw food with size depending on weight
    food_size = 10 + (food_weight - 1) * 5  # Bigger food for higher weight
    pygame.draw.rect(game_window, white, pygame.Rect(food_pos[0],food_pos[1], food_size, food_size))

    # Display score and level
    show_score_level()

    # Update display
    pygame.display.update()

    # Set game speed
    fps_controller.tick(difficulty)

pygame.quit()
print("Game exited successfully.")
```

```python
import pygame
import math

# initialize pygame
pygame.init()

# set up display
WIDTH, HEIGHT = 640, 480
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Drawing Tool")

# colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# variables
drawing = False
last_pos = None
color = BLUE
radius = 5
mode = "free_draw"  # free_draw, rectangle, circle, eraser, square, right_triangle, equilateral_triangle, rhombus
shapes = []  # stores drawn shapes

clock = pygame.time.Clock()

running = True
while running:
    screen.fill(BLACK)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        # handle keypresses for mode selection
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_r:
                color = RED
            elif event.key == pygame.K_g:
                color = GREEN
            elif event.key == pygame.K_b:
                color = BLUE
            elif event.key == pygame.K_e:
                mode = "eraser"
            elif event.key == pygame.K_f:
                mode = "free_draw"
            elif event.key == pygame.K_t:
                mode = "rectangle"
            elif event.key == pygame.K_c:
                mode = "circle"
            elif event.key == pygame.K_k:
                mode = "square"
            elif event.key == pygame.K_p:
                mode = "right_triangle"
            elif event.key == pygame.K_y:
                mode = "equilateral_triangle"
            elif event.key == pygame.K_d:
                mode = "rhombus"
            elif event.key == pygame.K_x:  # clear screen
                shapes.clear()
                screen.fill(BLACK)
        
        # handle mouse events
        if event.type == pygame.MOUSEBUTTONDOWN:
            drawing = True
            last_pos = event.pos
            if mode in ["rectangle", "circle", "square", "right_triangle", "equilateral_triangle", "rhombus"]:
                shapes.append((mode, color, last_pos, last_pos))
        
        if event.type == pygame.MOUSEBUTTONUP:
            drawing = False
            last_pos = None
        
        if event.type == pygame.MOUSEMOTION and drawing:
            if mode == "free_draw":
                pygame.draw.circle(screen, color, event.pos, radius)
                shapes.append(("dot", color, event.pos, radius))
            elif mode == "eraser":
                pygame.draw.circle(screen, BLACK, event.pos, radius)
                shapes.append(("dot", BLACK, event.pos, radius))
            elif mode in ["rectangle", "circle", "square", "right_triangle", "equilateral_triangle", "rhombus"]:
                shapes[-1] = (mode, color, shapes[-1][2], event.pos)
    
    # draw stored shapes
    for shape in shapes:
        shape_mode, shape_color, start_pos, end_pos = shape
        if shape_mode == "rectangle":
            pygame.draw.rect(screen, shape_color, (*start_pos, end_pos[0] - start_pos[0], end_pos[1] - start_pos[1]), 2)
        elif shape_mode == "square":
            side = min(abs(end_pos[0] - start_pos[0]), abs(end_pos[1] - start_pos[1]))
            pygame.draw.rect(screen, shape_color, (*start_pos, side, side), 2)
        elif shape_mode == "circle":
            center = ((start_pos[0] + end_pos[0]) // 2, (start_pos[1] + end_pos[1]) // 2)
            radius = max(abs(end_pos[0] - start_pos[0]), abs(end_pos[1] - start_pos[1])) // 2
            pygame.draw.circle(screen, shape_color, center, radius, 2)
        elif shape_mode == "right_triangle":
            pygame.draw.polygon(screen, shape_color, [
                start_pos,
                (end_pos[0], start_pos[1]),
                (start_pos[0], end_pos[1])
            ], 2)
        elif shape_mode == "equilateral_triangle":
            side_length = abs(end_pos[0] - start_pos[0])
            height = (math.sqrt(3) / 2) * side_length
            pygame.draw.polygon(screen, shape_color, [
                (start_pos[0], start_pos[1] + height),
                (start_pos[0] + side_length, start_pos[1] + height),
                (start_pos[0] + side_length // 2, start_pos[1])
            ], 2)
        elif shape_mode == "rhombus":
            dx = abs(end_pos[0] - start_pos[0]) // 2
            dy = abs(end_pos[1] - start_pos[1]) // 2
            center = (start_pos[0] + dx, start_pos[1] + dy)
            pygame.draw.polygon(screen, shape_color, [
                (center[0], start_pos[1]),
                (start_pos[0], center[1]),
                (center[0], end_pos[1]),
                (end_pos[0], center[1])
            ], 2)
        elif shape_mode == "dot":
            pygame.draw.circle(screen, shape_color, start_pos, end_pos)

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
